```python
import pandas as pd

df_ufo = pd.read_csv("data/ufo_sightings.csv", low_memory=False)
df_ufo
```

```python
df_ufo = df_ufo.dropna(axis=0)
df_ufo
```

```python
df_ufo["country"]
```

```python
df_ufo.groupby("country")["country"].count()
```

```python
df_ufo = df_ufo[df_ufo.country == "us"]
df_ufo
```

```python
df_ufo = df_ufo.drop(columns=["city", "duration (hours/min)", "country", "comments", "date posted"])
df_ufo
```

```python
df_ufo = df_ufo.rename(columns={"duration (seconds)":"duration"})
df_ufo
```

```python
df_ufo.dtypes

df_ufo["latitude"] = df_ufo["latitude"].apply(pd.to_numeric)
df_ufo.dtypes
```

```python
df_ufo.plot.scatter(x="longitude", y="latitude")
```

```python
df_ufo
```

```python
df_state = pd.read_csv("data/state-names.csv")
df_state
```

```python
df_ufo["state"] = df_ufo["state"].apply(lambda x:x.upper())
df_ufo
```

```python
df_ufo = df_ufo.merge(df_state, left_on="state", right_on="code")
df_ufo
```

```python
df_ufo["location"] = df_ufo.apply(lambda row: str(row.latitude)+','+str(row.longitude), axis=1)
```

```python
df_ufo = df_ufo.drop(columns=["latitude", "longitude", "code"])
df_ufo
```

```python
df_ufo.to_csv("output/ufos.csv", index=False)
```
